```python
from transformers import pipeline

# Load the GPT-2 model and text generation pipeline
model_name = "openai-community/gpt2"
text_generator = pipeline("text-generation", model=model_name)
```

```python
import gradio as gr

# Function to generate text from GPT-2 based on user input
def generate_text(prompt):
    return text_generator(prompt, max_length=100)[0]['generated_text']

# Gradio interface
gr.Interface(fn=generate_text,
             inputs="text",
             outputs="text",
             title="SDG Text Generator using GPT-2").launch()
```
